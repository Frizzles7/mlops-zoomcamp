```python
import pandas as pd
import pickle
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.feature_extraction import DictVectorizer
from sklearn.linear_model import LinearRegression, Lasso, Ridge
from sklearn.metrics import mean_squared_error
```

```python
# read parquet files
df = pd.read_parquet('./data/green_tripdata_2021-01.parquet')
```

```python
df.head()
```

```python
# create column of duration of trip in minutes, as dropoff time minus pickup time
df['duration'] = df.lpep_dropoff_datetime - df.lpep_pickup_datetime
df.duration = df.duration.apply(lambda td: td.total_seconds() / 60)
```

```python
sns.distplot(df.duration);
```

```python
df.duration.describe(percentiles=[0.95, 0.98, 0.99])
```

```python
df = df[(df.duration >= 1) & (df.duration <= 60)]
```

```python
categorical = ['PULocationID', 'DOLocationID']
numerical = ['trip_distance']
```

```python
# convert categoricals to strings
df[categorical] = df[categorical].astype(str)
```

```python
train_dicts = df[categorical + numerical].to_dict(orient='records')
dv = DictVectorizer()
X_train = dv.fit_transform(train_dicts)
```

```python
target = 'duration'
y_train = df[target].values
```

```python
lr = LinearRegression()
lr.fit(X_train, y_train)
```

```python
y_pred = lr.predict(X_train)
rmse = mean_squared_error(y_train, y_pred, squared=False)
rmse
```

```python
sns.distplot(y_pred, label='prediction')
sns.distplot(y_train, label='actual')
plt.legend();
```

```python
# create function to read and process data
def read_dataframe(filename):
    df = pd.read_parquet(filename)

    df['duration'] = df.lpep_dropoff_datetime - df.lpep_pickup_datetime
    df.duration = df.duration.apply(lambda td: td.total_seconds() / 60)

    df = df[(df.duration >= 1) & (df.duration <= 60)]

    categorical = ['PULocationID', 'DOLocationID']
    df[categorical] = df[categorical].astype(str)

    return df
    
```

```python
# use function to create training and validation
df_train = read_dataframe('./data/green_tripdata_2021-01.parquet')
df_val = read_dataframe('./data/green_tripdata_2021-02.parquet')
```

```python
len(df_train), len(df_val)
```

```python
# create pickup_dropoff
df_train['PU_DO'] = df_train['PULocationID'] + '_' + df_train['DOLocationID']
df_val['PU_DO'] = df_val['PULocationID'] + '_' + df_val['DOLocationID']
```

```python
categorical = ['PU_DO']
numerical = ['trip_distance']

train_dicts = df_train[categorical + numerical].to_dict(orient='records')
dv = DictVectorizer()
X_train = dv.fit_transform(train_dicts)

val_dicts = df_val[categorical + numerical].to_dict(orient='records')
X_val = dv.transform(val_dicts)
```

```python
target = 'duration'
y_train = df_train[target].values
y_val = df_val[target].values
```

```python
lr = LinearRegression()
lr.fit(X_train, y_train)
y_pred = lr.predict(X_val)
mean_squared_error(y_val, y_pred, squared=False)
```

```python
# save the dictvectorizer and the linear regression model
with open('./models/lin_reg.bin', 'wb') as f_out:
    pickle.dump((dv, lr), f_out)
```

```python
lass = Lasso(0.01)
lass.fit(X_train, y_train)
y_pred = lass.predict(X_val)
mean_squared_error(y_val, y_pred, squared=False)
```
